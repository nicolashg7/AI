```python
import random
from tabulate import tabulate
from transformers import pipeline

month_menu = []
month_ingredients = {}
previous_week = []

for i in range(4):

  weekly_menu = []

  legumes = [recipe for recipe in dic_All["Legumes"]["Recipes"] if recipe not in previous_week]
  meat = [recipe for recipe in dic_All["Meat"]["Recipes"] if recipe not in previous_week]
  salads = [recipe for recipe in dic_All["Salads"]["Recipes"] if recipe not in previous_week]
  fish = [recipe for recipe in dic_All["Fish"]["Recipes"] if recipe not in previous_week]

  weekly_menu.extend(random.sample(legumes, 2))
  weekly_menu.extend(random.sample(meat, 1))
  weekly_menu.extend(random.sample(salads, 1))
  weekly_menu.extend(random.sample(fish, 1))
  random.shuffle(weekly_menu)
  month_menu.append(weekly_menu)
  weekly_ingredients = []
  for dish in weekly_menu:
    weekly_ingredients.extend(All_ingredients[dish])
  weekly_ingredients = list(set(weekly_ingredients))

  month_ingredients[f"Week {i+1}"] = weekly_ingredients
  previous_week = weekly_menu.copy()

month_menu_with_weeks = [[f"Week {i+1}"] + menu for i, menu in enumerate(month_menu)]
print(tabulate(month_menu_with_weeks, tablefmt="grid"))
print("\n\n")

month_ingredients_with_weeks = [(week, ", ".join(ingredients)) for week, ingredients in month_ingredients.items()]
print(tabulate(month_ingredients_with_weeks, tablefmt="grid"))

prompt = f"""
Generate a description for the following weekly menu: {', '.join(month_menu[0])}.
After that, provide a detailed shopping list for the ingredients needed: {', '.join(month_ingredients["Week 1"])}.
"""

print(prompt)


# Use GPT-Neo for more coherence or continue with GPT-2 if preferred
pipe = pipeline("text-generation", model="gpt2")

prompt = f"""
You are a helpful assistant.

### Step 1: Create a description for the following weekly menu:
{', '.join(month_menu[0])}.

### Step 2: Now, provide a detailed shopping list for the following ingredients:
{', '.join(month_ingredients["Week 1"])}.
"""

output = pipe(prompt, max_length=500, do_sample=True, temperature=0.7)

# Print the generated text
print(output[0]["generated_text"])


#create a text with AI to explain the menu for each week and detail the shopping list


```
